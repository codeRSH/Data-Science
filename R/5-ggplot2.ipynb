```r
## Grammar of Graphics ##

# Following are the components of Grammer of Graphics : 
# Data
# Aesthetics
# Geometries
# Statistics
# Facets
# Coordinates
# Themes


getwd()

movies <- read.csv("P2-Movie-Ratings.csv")
head(movies)
colnames(movies)

colnames(movies) <- c("Film", "Genre", "Critic.Ratings", "Audience.Ratings", "Budget", "Year")

colnames(movies)

tail(movies)
str(movies)
summary(movies)

# Convert Movies (Integer) into Factors i.e. something having Levels
movies$Year <- factor(movies$Year)

str(movies)
summary(movies)



# ----------- Aesthetics

library(ggplot2)
ggplot(data=movies, aes(x=Critic.Ratings, y = Audience.Ratings))

# Add Geometry
ggplot(data = movies, aes(x = Critic.Ratings, y = Audience.Ratings)) +
geom_point()

# Add Color
ggplot(data = movies, aes(x = Critic.Ratings, y = Audience.Ratings, 
       color = Genre)) +
geom_point()


# Add Size
ggplot(data = movies, aes(x = Critic.Ratings, y = Audience.Ratings, 
       color = Genre, size = Genre)) +
geom_point()


# Add Size - Better Way
ggplot(data = movies, aes(x = Critic.Ratings, y = Audience.Ratings, 
       color = Genre, size = Budget)) +
geom_point()



# -------- Plotting with Layers
# ggplot inface creates an object which can be assigned to a variable
p <- ggplot(data = movies, aes(x = Critic.Ratings, y = Audience.Ratings, 
       color = Genre, size = Budget))

p

# Point
p + geom_point()


# Lines
p + geom_line()

# Point + Line
p + geom_point() + geom_line()
p + geom_line() + geom_point()




# ------- Overriding Aesthetics

q <- ggplot(data = movies, aes(x = Critic.Ratings, y = Audience.Ratings,
             color = Genre, size = Budget))

# Add Geometric Layer
q + geom_point()

# Override Aes
q + geom_point(aes(size = Critic.Ratings))

q + geom_point(aes(color = Budget))

# Value of q itself doesn't change by override aes
q + geom_point()

q + geom_point(aes(x = Budget)) + xlab("Budget $$")

q + geom_line() + geom_point()
q + geom_line(size = 1) + geom_point()



# ------- Mapping Vs Setting

r <- ggplot(data = movies, aes(x = Critic.Ratings, y = Audience.Ratings))

# Set Color
# 1. Mapping (done this way till now)
r + geom_point(aes(color = Genre))
# 2. Setting
r + geom_point(color = "DarkGreen")

# Error:
# r + geom_point(aes(color = "DarkGreen"))



# Set Size
# 1. Mapping
r + geom_point(aes(size = Budget))
# 2. Setting
r + geom_point(size = 5)

# Error
# r + geom_point(aes(size = 200))


#------- Histogram and Density Charts

s <- ggplot(data = movies, aes(x = Budget))
#s <- ggplot(data = movies[movies$Genre == "Drama", ], aes(x = Budget))
s + geom_histogram(binwidth = 10)

# Add Color (here we use Fill argument)
s + geom_histogram(binwidth = 10, aes(fill = Genre))

# Add border (Color argument is used for Border color)
s + geom_histogram(binwidth = 10, aes(fill = Genre), color = "Black")

# Sometimes we create Density Chart
s + geom_density()
s + geom_density(aes(fill = Genre))
s + geom_density(aes(fill = Genre), position = "stack")



# ---- Starting Layer Tips
t <- ggplot(data = movies, aes(x = Audience.Ratings))
t + geom_histogram(binwidth = 10, fill = "White", color = "blue" )


# Another way
t <- ggplot(data = movies)
t + geom_histogram(binwidth = 10, aes(x = Audience.Ratings),
                    fill = "white", color = "Blue")

t + geom_histogram(binwidth = 10, aes(x = Critic.Ratings),
                    fill = "white", color = "Blue")

# Yet another way
# Creating an empty plot and then modifying through layers as necessary
t <- ggplot()


# ------ Statistical Transformations

?geom_smooth
u <- ggplot(data = movies, aes(x = Critic.Ratings, 
            y = Audience.Ratings, color = Genre))
u + geom_point()

u + geom_point() + geom_smooth()
u + geom_point() + geom_smooth(fill = NA)


# Boxplot
u <- ggplot(data = movies, aes(x = Genre, 
            y = Audience.Ratings, color = Genre))
u + geom_boxplot()
u + geom_boxplot(size = 1.2)

u + geom_boxplot() + geom_point()

# tip : use jitter instead of point with boxplot
u + geom_boxplot() + geom_jitter()

# another way
u + geom_jitter() + geom_boxplot()
u + geom_jitter() + geom_boxplot(size = 1.2, alpha = 0.5)


n <- ggplot(data = movies, aes(x = Genre, y = Critic.Ratings, color = Genre))
n + geom_point()
n + geom_jitter()
n + geom_jitter() + geom_boxplot(size = 1.2, alpha = 0.6)






# ----- Using Facets
v <- ggplot(data = movies, aes(x = Budget))
v + geom_histogram(binwidth = 10, aes(fill = Genre), color = "Black")

# Facets
v + geom_histogram(binwidth = 10, aes(fill = Genre), color = "black") + 
    facet_grid(Genre~., scales = "free")


# Scatter plot
w <- ggplot(data = movies, aes(x = Critic.Ratings, y = Audience.Ratings,
            color = Genre))    
w + geom_point(size = 3)

# Facets
w + geom_point(size = 3) + facet_grid(Genre~.)
w + geom_point(size = 3) + facet_grid(.~Year)
w + geom_point(size = 3) + facet_grid(Genre~Year) + geom_smooth(fill = NA)


w + geom_point(aes(size = Budget)) + geom_smooth() + 
       facet_grid(Genre~Year)



# ------ Coordinates

# Limits 
m <- ggplot(data = movies, aes(x = Critic.Ratings, y = Audience.Ratings,
               color = Genre))
m + geom_point()

# Limit basically discards the data which doesn't confirm to x or y axis
m + geom_point() + xlim(50,100)
m + geom_point() + xlim(50, 100) + ylim(50, 100)

# Limits doesn't always work well. See this
o <- ggplot(data = movies, aes(x = Budget))
o + geom_histogram(binwidth = 10, aes(fill = Genre), color = "Black")
o + geom_histogram(binwidth = 10, aes(fill = Genre), color = "Black") + 
       ylim(0, 50)

# Instead zoom : 
o + geom_histogram(binwidth = 10, aes(fill = Genre), color = "black") +
       coord_cartesian(ylim = c(0,50))



w + geom_point(aes(size = Budget)) + geom_smooth() + 
       facet_grid(Genre~Year)  + coord_cartesian(ylim = c(0,100))



#----- Theme
i <- ggplot(data = movies, aes(x = Budget))
j <- i + geom_histogram(binwidth = 10, aes(fill = Genre), color = "Black")
j


# Axes Labelling
j + xlab("Money Axis") + ylab("Number of Movies")

# Label Formatting
j + xlab("Money Axis") + ylab("Number of Movies") + 
       theme(axis.title.x = element_text(color = "DarkGreen", size = 30),
              axis.title.y = element_text(color = "Red", size = 30))


# Tick Mark Formatting
j + xlab("Money") + ylab("Movie No.") +
  theme(axis.title.x = element_text(color = "Green", size = 20),
        axis.title.y = element_text(color = "Blue", size = 25), 
        axis.text.x = element_text(size = 15),
        axis.text.y = element_text(size = 15))  

# Legend Formatting
j + xlab("Money") + ylab("Movie count") +
       theme(axis.title.x = element_text(color = "Green", size = 20),
       axis.title.y = element_text(color = "blue", size = 20),
       axis.text.x = element_text(size = 15),
       axis.text.y = element_text(size = 15),
       legend.title = element_text(size = 15),
       legend.text = element_text(size = 10),
       legend.position = c(1,1),
       legend.justification = c(1,1)
       )

# Title
# Loading and Mapping font is important here
windowsFonts(Courier=windowsFont("Courier"))

j + xlab("Money") + ylab("Movie count") +
       ggtitle("Movie Distribution") +
       theme(axis.title.x = element_text(color = "Green", size = 20),
       axis.title.y = element_text(color = "blue", size = 20),
       axis.text.x = element_text(size = 15),
       axis.text.y = element_text(size = 15),
       legend.title = element_text(size = 15),
       legend.text = element_text(size = 10),
       legend.position = c(1,1),
       legend.justification = c(1,1),
       plot.title = element_text(color = "DarkBlue", size = 40, 
                     family = "Courier")
       )
```
