## ------- $ Sign -------

$ Sign doesn't work for matrices but only for data frames. 
In data frame only columns are named the rows are not named (unlike matrices)

```r
# stats
# head(stats)

# Accessing element of Data Frame
stats[3, 4] 
```

```r
# Access by index and name 

stats[3, "Internet.users"]
```

```r
# Access whole column "Income.Group"
stats[,"Income.Group"]
```

```r
# Access Column using $ 

stats$Income.Group
```

```r
# Access a Particular element of the selected column

stats$Internet.users[4]
```

## Basic Operations

```r
# Subsetting

stats[3:10,]   
```

```r
stats[c(2, 93),]
```

```r
# [] operation on data frame

# Selecting First Row
stats[1, ]

is.data.frame(stats[1, ])   # No need for drop=FALSE for row
```

```r
# Selection of 1st Column

stats[, 1]
is.data.frame(stats[,1]) # Not a Data Frame! Becomes a Vector now because R finds that convenient
```

```r
# When using drop=FALSE it will not drop a dimension even if it becomes a vector

stats[,1, drop = FALSE]
is.data.frame(stats[,1, drop = FALSE])
```

```r
# Multiple columns

head(stats)
stats$Birth.rate * stats$Internet.users
```

```r
# Add a new columns to data frame

stats$MyCol <- stats$Birth.rate + stats$Internet.users
head(stats)
```

```r
# Values of a vector with length lower than the rows in data frame will get repeated. 

nrow(stats)
stats$xyz <- 1:5 # This MUST be a multiple of existing rows else it will give error. 
head(stats, n = 12)
```

```r
# Remove Columns

stats$MyCol <- NULL
stats$xyz <- NULL

head(stats)
```

```r
# Filtering Data Frame

head(stats)

filter <- stats$Internet.users < 2 # Creates a vector with TRUE/FALSE values
stats[filter, ]   # Only rows with filter containing TRUE will be returned
```

```r
stats[stats$Birth.rate > 40, ]
```

```r

stats[stats$Birth.rate > 40 & stats$Internet.users < 2, ]
```

```r

stats[stats$Income.Group == "High income", ]
```

## Merging Data Frames

```r
# Merging 2 Data frame stats and mydf using common key Country.Code

merged <- merge(stats, mydf, by.x =  "Country.Code", by.y = "Country.Code")
head(merged)
```

```r
# Dropping an extra column which is not required

merged$Country.Name.y <- NULL
tail(merged)
```

```r
# Rename Column Country.Name.x to Country.Name

colnames(merged)[2] <- "Country.Name"
str(merged)
```

```r
# Some more qplots on merged Data Frame now

qplot(data=merged, x=Birth.rate, y=Internet.users, size=I(4), color=Region)
```

```r
# Shapes

qplot(data=merged, y=Birth.rate, x=Internet.users, size=I(3), 
      shape=I(15), color=Region)
```

```r
# Transparency

qplot(data=merged, y=Birth.rate, x=Internet.users, size=I(3), 
      shape=I(19), alpha=I(0.5), color=Region)
```

```r
# Title
qplot(data=merged, y=Birth.rate, x=Internet.users, 
      color=Region, size=I(3), 
      shape=I(19), alpha=I(0.5),
      main = "Internet Users Vs Birth Rate")
```
