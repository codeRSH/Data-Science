## ------- $ Sign -------

$ Sign doesn't work for matrices but only for data frames. 
In data frame only columns are named the rows are not named (unlike matrices)

```r
# stats
# head(stats)

# Accessing element of Data Frame
stats[3, 4] 
```

```r
# Access by index and name 

stats[3, "Internet.users"]
```

```r
# Access whole column "Income.Group"
stats[,"Income.Group"]
```

```r
# Access Column using $ 

stats$Income.Group
```

```r
# Access a Particular element of the selected column

stats$Internet.users[4]
```

```r
# Get all the factors of the column 

levels(stats$Income.Group)
```

```r
stats[c(2, 93),]
```

```r
# [] operation on data frame

# Selecting First Row
stats[1, ]

is.data.frame(stats[1, ])   # No need for drop=FALSE for row
```

```r
# Selection of 1st Column

stats[, 1]
is.data.frame(stats[,1]) # Not a Data Frame! Becomes a Vector now because R finds that convenient
```

```r
# When using drop=FALSE it will not drop a dimension even if it becomes a vector

stats[,1, drop = FALSE]
is.data.frame(stats[,1, drop = FALSE])
```
